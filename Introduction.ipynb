## Introduction

### Problem Background

New York City remains to be the financial capital of the world since many years. Businesses have been established and flourished in this multicultural city. It houses various industries such as banking and finance, retail, world trade, transportation, tourism, real estate, digital media, traditional media, advertising, legal services, accountancy, insurance, entertainment and  many more. It has a diverse population with people from varied backgrounds.

More the options, more is the competition. Thus, businesses need to work hard to survive in the ever growing market of New York City. The investment of a business is also high as NYC is a developed city. Subsequently, any new business wanting to start its journey should have an in-depth analysis of the market positionn. An informed decision helps go a long way in running the business.

### Problem Description

New York is house to numerous multi-cuisine restaurants. The credit is shared among the immigrants who brought with them flavors of their homeland. Some examples include:

1. Central and European immigrants - bagels, cheesecake, hot dogs
2. Italian immigrants - pizza and italian cuisine
3. Irish immigrants - pastrami and corned beef
4. Asian immigrants - sandwich and spices
5. Middle Eastern immigrants - falafel, kebab

This hints to the background study required to set up a business in order for it to run successfully. Some of the factors which affect are:

1. Population of the city
2. Demographic location
3. Nearby markets and wholesale dealers
4. Competitors
